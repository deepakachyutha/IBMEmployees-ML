```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

employee = pd.read_csv(r'..\Dataset\IBM-HR-Employee-Attrition-Encoded.csv')
```

```python
employee.groupby('Department')['Attrition'].mean().plot(kind='bar')
plt.title("Attrition Rate by Department")
plt.xlabel("Department")
plt.ylabel("Attrition Rate")
plt.show()
```

```python
sns.countplot(employee, x = "Gender", hue='Attrition')
plt.title("Attrition by Gender")
plt.ylabel("Count")
plt.show()

employee['Age'].hist(bins=20)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("No of Employees")
plt.show()

employee['DistanceFromHome'].hist(bins=20)
plt.title("Distance from Home Distribution")
plt.xlabel("Distance from Home")
plt.ylabel("Kilometers")
plt.show()

employee['MonthlyIncome'].hist(bins=20)
plt.title("Monthly Income Distribution")
plt.xlabel("Monthly Income")
plt.ylabel("No of Employees")
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(employee.corr(), annot=False, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
sns.boxplot(employee, x='JobRole', y='MonthlyIncome')
plt.title("MonthlyIncome by JobRole")
plt.show()
```
